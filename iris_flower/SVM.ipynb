```python
data = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.head()
```

```python
Y_column = 'species'
X_columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
```

```python
train_data, test_data = train_test_split(data, test_size = 0.2)
```

```python
X_train, X_test, Y_train, Y_test = train_data[X_columns], test_data[X_columns], train_data[Y_column], test_data[Y_column]
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussianNB = GaussianNB()
```

```python
gaussianNB.fit(X_train,Y_train)
```

```python
predicted_gaussianNB = gaussianNB.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
gaussiannb_acc = accuracy_score(Y_test, predicted_gaussianNB)
```

```python
print('Performance of Naive Bayes is %.3f', gaussiannb_acc)
```

```python
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression()
```

```python
logReg.fit(X_train,Y_train)
predicted_logreg = logReg.predict(X_test)
logreg_acc = accuracy_score(Y_test,predicted_logreg)
```

```python
print('Performance of Logistic regression is %.3f', logreg_acc)
```

```python
from sklearn.linear_model import SGDClassifier

linear_svm = SGDClassifier(loss = 'hinge')
linear_svm.fit(X_train,Y_train)
predicted_linearsvm = linear_svm.predict(X_test)
linearsvm_acc = accuracy_score(Y_test,predicted_linearsvm)
```

```python
print('Performance of Linear SVM is %.3f', linearsvm_acc)
```

```python
from sklearn.svm import SVC

linear_svm_2 = SVC(kernel = 'linear')
linear_svm_2.fit(X_train,Y_train)
predicted_linearsvm2 = linear_svm_2.predict(X_test)
linearsvm2_acc = accuracy_score(Y_test,predicted_linearsvm2)
```

```python
print('Performance of Linear SVM using SVC is %.3f', linearsvm2_acc)
```

```python
poly_svm = SVC(kernel = 'poly')
poly_svm.fit(X_train,Y_train)
predicted_polysvm = poly_svm.predict(X_test)
polysvm_acc = accuracy_score(Y_test,predicted_polysvm)
```

```python
print('Performance of Polynomial SVM is %.3f', polysvm_acc)
```

```python
rbf_svm = SVC(kernel = 'rbf',probability = True)
rbf_svm.fit(X_train,Y_train)
predicted_rbfsvm = rbf_svm.predict(X_test)
rbfsvm_acc = accuracy_score(Y_test,predicted_rbfsvm)
```

```python
print('Performance of rbf SVM is %.3f', rbfsvm_acc)
```
